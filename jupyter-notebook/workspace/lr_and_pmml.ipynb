```python
train_sample = pd.DataFrame(np.random.rand(100, 6))
train_sample.columns = train_sample.columns.astype(str)
train_sample = train_sample.rename({c:"var"+c for c in train_sample.columns}, axis="columns")
train_sample.to_csv("train_sample.csv")
train_target = pd.DataFrame(np.random.randint(2, size=(100, 1)))
train_sample.join(train_target)
```

```python
classifier = LogisticRegression()
classifier.fit(train_sample, train_target.values.flatten())
```

```python
classifier.predict(train_sample)
```

```python
# from sklearn_pandas import DataFrameMapper
# default_mapper = DataFrameMapper([(i, None) for i in train_sample.columns + ['Species']])
from sklearn2pmml import PMMLPipeline
pipeline = PMMLPipeline([
  ("the-classifier", classifier)
])

sklearn2pmml.sklearn2pmml(pipeline, "the-classifier.pmml")
```
